```python
import findspark
from functools import reduce
from pyspark import SparkContext
from pyspark.sql import SparkSession, Window, Row
from pyspark.sql.functions import *
from pyspark.sql.types import *
import plotly as py
import matplotlib.pyplot as plt

import pyspark.pandas as ps
```

```python
#https://spark.apache.org/docs/latest/sql-getting-started.html
spark = SparkSession \
        .builder \
        .appName("firstSpark") \
        .getOrCreate()
```

```python
# SPARK
sdf = spark.read.options(inferSchema='True',
              header='True').csv('./Data/Matches.csv')
# PANDAS-ON-SPARK
pdf = ps.read_csv('./Data/Matches.csv')
```

```python
sdf.limit(5).show()
```

```python
pdf.head()
```

```python
#lets rename some of the columns of the sdf
old_cols = sdf.columns[-3:]
new_cols = ["HomeTeamGoals", "AwayTeamGoals", "FinalResult"]
old_new_cols = [*zip(old_cols, new_cols)]
for old_col, new_col in old_new_cols:
    sdf = sdf.withColumnRenamed(old_col, new_col)
    
sdf.limit(5).toPandas()
```

```python
#lets rename some of the columns of the pdf
old_cols = pdf.columns[-3:]
new_cols = ["HomeTeamGoals", "AwayTeamGoals", "FinalResult"]
old_new_cols = [*zip(old_cols, new_cols)]
for old_col, new_col in old_new_cols:
    pdf = pdf.rename(columns = {old_col: new_col})
pdf.head()
```
